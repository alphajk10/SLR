```python
from sklearn.impute import SimpleImputer
X = df[["Experience_Years","Salary"]]
y = df["Department"]
```

```python
print("--- Original X and y ---")
print(X)
print(y)
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
x = imputer.fit_transform(X)
```

```python
imputer_sc = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
y = imputer_sc.fit_transform(y.values.reshape(-1,1))
```

```python
df[["Experience_Years","Salary"]] = x
y = y.ravel()
df["Department"] = y
```

```python
from sklearn.ensemble import IsolationForest

iso = IsolationForest(
    n_estimators=100,
    contamination= 'auto',
    random_state=42
)

df['Outlier'] = iso.fit_predict(X)
```

```python
df[df['Outlier'] == -1]
```

```python
dataset_fixed = df[df['Outlier'] == 1]
```

```python
dataset_fixed = dataset_fixed.drop(columns=['Outlier'])
```

```python
x1 = dataset_fixed.iloc[:, 1:2]
y1 = dataset_fixed.iloc[:, 2:3]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x1,y1,test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```
